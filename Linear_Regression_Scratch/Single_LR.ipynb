```python
# MSE
def loss_function(m, b, data):
    error = 0
    for i in range(len(data)):
        y = data.iloc[i]['Salary']
        x = data.iloc[i]['YearsExperience']
        error += (y - (m * x + b)) ** 2
    error = error / float(len(data))
    return error
```

```python
def gradient_descent(m_now, b_now, data, alpha):
    m_gradient = 0
    b_gradient = 0
    n = len(data)

    for i in range(n):
        y = data.iloc[i]['Salary']
        x = data.iloc[i]['YearsExperience']

        m_gradient += -(2 / n) * (y - (m_now * x + b_now)) * x
        b_gradient += -(2 / n) * (y - (m_now * x + b_now))
    
    m_update = m_now - alpha * m_gradient
    b_update = b_now - alpha * b_gradient

    return m_update, b_update
```

```python
m = 0 
b = 0
alpha = 0.001
iterations = 1000
```

```python
for i in range(iterations):
    m, b = gradient_descent(m, b, data, alpha)
```
